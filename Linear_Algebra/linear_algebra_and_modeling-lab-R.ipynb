### 11. Fit a function to Random Data

#### Generate two random points.

```r
set.seed(10)
p1 <- runif(2); p2 <- runif(2)
p1
p2
```

#### Plot these two points

```r
p1
p2

options(repr.plot.width=12, repr.plot.height=6)
plot(NA, xlim=c(-1,1), ylim=c(-1,1), xlab="X", ylab="Y")
abline(h=0) 
abline(v=0)
points(0.50747820315882,0.306768506066874, col="red")
points(0.426907666493207,0.693102080840617, col="blue") 
```

Use linear algebra to fit a line to these two points.

We know that the equation looks like this

$$y_i=\beta_0+\beta_1x_i$$

Which we can rewrite as

$$(1,x_i)^T(\beta_0, \beta_1)=y_i$$

or

$$\left(\begin{matrix}1&x_i\end{matrix}\right)\left(\begin{matrix}\beta_0\\ \beta_1\end{matrix}\right)=y_i$$

We can use this to make

$$X\beta=\mathbf{y}$$

```r
X = rbind(c(1, 0.50747820315882), c(1, 0.426907666493207))
y = c(0.306768506066874, 0.693102080840617)
X
y
```

### Find $\beta$

```r
beta = solve(X,y)
beta
```

```r
options(repr.plot.width=12, repr.plot.height=6)
plot(NA, xlim=c(-4,4), ylim=c(-3,5), xlab="X", ylab="Y")
abline(h=0); abline(v=0)
xx <- seq(-3,3,by = 0.1)
points(0.50747820315882,0.306768506066874, col="red")
points(0.426907666493207,0.693102080840617, col="blue") 
lines(xx, beta[1] + beta[2]*xx, col='black')
```

Plot the two points and the line described by $\beta$.

### 12. Find the line of best fit

```r
beta_0 <- 2
beta_1 <- -3
xx <- seq(-5,5,by=.1)
n <- length(xx)
noise <- 5*runif(n)
yy = beta_0 + beta_1*xx + noise
```

```r
plot(xx,yy)
```

#### Problem set up

Create your matrices and vectors so that

$$X\beta=\mathbf{y}$$

Solve $$X^TX\beta=X^T\mathbf{y}$$

```r
#y = -3x + 2
X = rbind(c(1, -4), c(1, -2), c(1, 2), c(1, 4))
y = c(14, 8, -4, -10)

XTX = t(X)%*%X
XTy = t(X)%*%y

beta = solve(XTX,XTy)
beta
```

#### Plot the line of best fit

```r
options(repr.plot.width=12, repr.plot.height=6)
plot(xx,yy, xlim=c(-5,5), ylim=c(-10,20), xlab="X", ylab="Y")
abline(h=0); abline(v=0)
xx <- seq(-5,5,by=.1)
lines(xx, (beta_0+2) + (beta_1)*xx, col='blue')
```
