```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
def calculate_r_2_features(data, feature):
    new_data = data.drop(feature, axis=1)
    
    (X_train, X_test, y_train, y_test) = train_test_split(new_data, data[feature])
    
    regressor = DecisionTreeRegressor()
    regressor.fit(X_train, y_train)
    score = regressor.score(X_test, y_test)
    return score
```

```python
def mean_for_r_2(data,feature):
    scores = []
    for _ in range(100):
        scores.append(calculate_r_2_features(data, feature))
        
    scores = np.array(scores)
    return scores.mean()
```

```python
def display_outliers(dataframe, col, params = 1.5):
    Q1 = np.percentile(dataframe[col], 25)
    Q3 = np.percentile(dataframe[col], 75)
    tukey_window = params*(Q3-Q1)
    less_than_Q1 = dataframe[col] < Q1 - tukey_window
    more_than_Q3 = dataframe[col] > Q3 + tukey_window
    tukey_mask = (less_than_Q1 | more_than_Q3)
    return dataframe[tukey_mask]
```
