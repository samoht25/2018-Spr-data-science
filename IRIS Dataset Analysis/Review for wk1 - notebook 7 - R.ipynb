```r
iris_df <- read.csv("data/iris.csv", row.names=1)
```

```r
head(iris_df)
```

```r
library(repr)
```

```r
options(repr.plot.width=20, repr.plot.height=10) #plt.figure(figsize(0,20))
```

```r
pairs(iris_df)
```

```r
library(ggplot2)
```

```r
options(repr.plot.width=20, repr.plot.height=5)
```

```r
ggplot(iris_df, aes(x=petal_length, y=petal_width)) +
       geom_point() + geom_smooth(method=lm)
```

```r
lm_1_var = lm(petal_width~petal_length, iris_df)
```

```r
lm_1_var
```

```r
options(repr.plot.width=20, repr.plot.height=5)
iris_df$petal_width_prediction_1_var = predict(lm_1_var)
```

```r
head(iris_df$petal_width_prediction_1_var)
```

```r
ggplot(iris_df, aes(petal_length, petal_width)) + geom_point() + geom_smooth(method=lm) +
geom_point(aes(petal_length, petal_width_prediction_1_var), color="red")
```

```r
lm_2_var = lm("petal_width ~ petal_length + sepal_length", iris_df)
```

```r
lm_2_var
```

```r
lm_3_var = lm("petal_width ~ petal_length + sepal_length + sepal_width", iris_df)
```

```r
lm_3_var
```

```r
options(repr.plot.width=20, repr.plot.height=5)
iris_df$petal_width_prediction_2_var = predict(lm_2_var)
iris_df$petal_width_prediction_3_var = predict(lm_3_var)

ggplot(iris_df, aes(petal_length, petal_width)) + geom_point() + geom_smooth(method=lm) +
geom_point(aes(petal_length, petal_width_prediction_1_var), color="red") + 
geom_point(aes(petal_length, petal_width_prediction_2_var), color="yellow") +
geom_point(aes(petal_length, petal_width_prediction_3_var), color="green")
```
